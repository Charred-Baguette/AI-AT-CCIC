# My Question

### Write your question here.

Can just packet length be used to indicate the likely severity level of an attack?

# My Analysis

```python
# Analyze here
#focus on frequency table
#histograms
#expected value
#standard deviation of value
#focus on a model

print('First we must understand the general frequencies for all severity levels. Therefore we should use a relative frequency table to determine the frequency of each severity. This can help us find their respective probabilities.')
HighSeverity=0
MediumSeverity=0
LowSeverity=0
UnknownSeverity=0

for i in df['Severity Level']:
    if i == 'High':
        HighSeverity+=1
    elif i == 'Medium':
        MediumSeverity+=1
    elif i == 'Low':
        LowSeverity+=1
    else:
        UnknownSeverity+=1
        print(i)
highfreq=HighSeverity/len(df)
medfreq=MediumSeverity/len(df)
lowfreq=LowSeverity/len(df)
unknownfreq=UnknownSeverity/len(df)
frequencies={'High':highfreq, 'Medium':medfreq, 'Low':lowfreq, 'Unknown':unknownfreq}
freqdf = pd.DataFrame(list(frequencies.items()), columns=['Severity', 'Relative Frequency'])
print(freqdf)
```

From the relative frequency table we can understand that each severity level has a roughly equal relative frequency. Therefore their relative probabilities are also likely similar.

```python
sns.histplot(data=df, x='Packet Length', bins=100)
```

From the histogram I can understand that each packet length has a roughly equal chance of occuring. The fact that this graph has a uniform distribution shows that each packet length has that roughly equal chance of occuring. This makes me nervous about the possibilities of using Packet Length to determine severity, but I continue none the less. 

```python

print('Using the frequencies found in the relative frequency table, I will find the expected value of the packet length for each severity level.')
SumHigh=0
SumMedium=0
SumLow=0

ExpectedHigh=0
ExpectedMedium=0
ExpectedLow=0

for index, row in df.iterrows():
    if row['Severity Level'] == 'High':
        tmp=row['Packet Length']*frequencies['High']
        ExpectedHigh+=tmp
    elif row['Severity Level']== 'Medium':
        tmp=row['Packet Length']*frequencies['Medium']
        ExpectedMedium+=tmp
    elif row['Severity Level'] == 'Low':
        tmp=row['Packet Length']*frequencies['Low']
        ExpectedLow+=tmp

print(f'Expected Packet Length for High Severity: {ExpectedHigh}')
print(f'Expected Packet Length for Medium Severity: {ExpectedMedium}')
print(f'Expected Packet Length for Low Severity: {ExpectedLow}')
```

These expected values give me quite a bit of insight. It appears as though Medium Severity threats are expected to use the most network resources. In the same way, Low Severity threats are expected to use the least amount of resources and can be less prioritized. From all of this data, it would be smart to understand the standard deviation of the expected lengths and graph the relative distribution of packet length for each severity level. Given the proximity of the threats, it appears to be less feasible to use a linear regression model as a means to predict severity level by packet length.

```python
print('proceeding to standard deviation...')

STDHigh=0
STDMedium=0
STDLow=0

SqDifHigh=0
SqDifMed=0
SqDifLow=0

for index, row in df.iterrows():
    if row['Severity Level'] == 'High':
        tmp=(row['Packet Length']-ExpectedHigh)**2
        SqDifHigh+=tmp
    elif row['Severity Level'] == 'Medium':
        tmp=(row['Packet Length']-ExpectedMedium)**2
        SqDifMed+=tmp
    elif row['Severity Level'] == 'Low':
        tmp=(row['Packet Length']-ExpectedLow)**2
        SqDifLow+=tmp

STDHigh=(SqDifHigh/len(df))**.5
STDMedium=(SqDifMed/len(df))**.5
STDLow=(SqDifLow/len(df))**.5
print(f'High Severity Standard Deviation: {STDHigh}')
print(f'Medium Severity Standard Deviation: {STDMedium}')
print(f'Low Severity Standard Deviation: {STDLow}')
```

These standard deviations are very worrisome. They are indicative of high variance meaning there is a lot of overlap with packet length for each severity level. In fact, this suggests that for each expected value, 68% of packet lengths should fall within one standard deviation above and below the expected value. Please note that this statement is not likely to be completely accurate as the data does not appear to have a normal distribution (look at the graph above). Therefore, it is not feasible to use packet length as a prediction metric for severity level. I do not need to graph out the histograms for packet length for each severity level nor create linear regression model as the standard deviations and expected values prove that packet length is not useful for predicting the severity level.
