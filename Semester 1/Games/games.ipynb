```python
codcount = df['name'].str.contains('Call of Duty', case=False, na=False).sum()
codplace=[]
for idx in df.index:
    name_value = df.at[idx, 'name']
    if pd.notna(name_value) and 'Call of Duty' in name_value:
        codplace.append(idx)
print(codcount)
print(codplace)
```

```python
for i in codplace:
    cod = df.loc[i]  
    print(cod)
```

```python
def scatter(df, x, y, outliers):
    idx=[]
    for i in range(outliers):
        idx=df[x].idxmax()
        df.drop([idx], inplace=True)
    sb.scatterplot(data=df, x=x, y=y)
scatter(df=df, x='metacritic_score', y='price', outliers=0)
```

```python
dfclean = df[df['metacritic_score'] != 0]
scatter(df=dfclean, x='metacritic_score', y='price', outliers=0)
```

There appears to be 2 major outliers at around price = 1000 and around price=500. It is interesting that some games are so expensive without a metacritic score. When the metacritic score is not 0 it can be seen that there is no real correlation between metacritic score and price.
